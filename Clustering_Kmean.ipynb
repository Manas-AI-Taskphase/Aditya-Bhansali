```python
from sklearn.cluster import KMeans
```

```python
def optimize_k(datas,max_k):
    means=[]
    inertias=[]
    for k in range(1,max_k):
        kmeans=KMeans(n_clusters=k)
        kmeans.fit(datas)
        means.append(k)
        inertias.append(kmeans.inertia_)
    fig=plt.subplots(figsize=(10,5))
    plt.plot(means,inertias,'o-')
    plt.xlabel("Number of clusters")
    plt.ylabel("Inertia")
    plt.grid(True)
    plt.show()
```

```python
X=pca_df[['PC1','PC2']]
X
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
def euclidean_distance(point, data):
    return np.sqrt(np.sum((point - data)**2, axis=1))
```

```python
def K_means(k, X, max_iterations=200):
    centroids = np.random.uniform(np.amin(X, axis=0), np.amax(X, axis=0), size=(k, X.shape[1]))
    for i in range(max_iterations):
        y=[]

        for data_point in X:
            distances = euclidean_distance(data_point, centroids)
            cluster_num = np.argmin(distances)
            y.append(cluster_num)

        y = np.array(y)

        cluster_indices = []
        
        for i in range(k):
            cluster_indices.append(np.argwhere(y == i)) 

        cluster_centers = []

        for i, indices in enumerate(cluster_indices):
            if len(indices) == 0:
                cluster_centers.append(centroids[i])
            else:
                cluster_centers.append(np.mean(X[indices], axis=0)[0])
        
        if np.max(centroids - np.array(cluster_centers)) < 0.0001:
            break
        else:
            centroids = np.array(cluster_centers)
    
    return y
```

```python
k = 4
max_iters = 300
cluster= K_means(k, X, max_iters)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster, cmap='viridis')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
